```python
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
dataset1
```

```python
plt.hist(dataset1) # ヒストグラム描写
```

```python
dataset2 = randn(80)
```

```python
plt.hist(dataset2, color = 'indianred')
```

```python
plt.hist(dataset1, normed=True)  # normed:標準化する
```

```python
plt.hist(dataset1, normed=True, alpha = 0.5, bins =20)   # alphaで透明度を指定
plt.hist(dataset2, normed=True, alpha = 0.5, bins = 20, color='indianred')
```

```python
data1 = randn(1000)
```
