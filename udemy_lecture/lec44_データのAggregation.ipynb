## Aggregationとは、大量のデータからなんらかの方法を使って、特徴的な量(平均・最大・最小等々)を取得する。

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
# 機械学習用のサンプルデータ
```

```python
!cat winequality-red.csv
```

```python
# セミコロン区切りのデータをpdで読み込む
dframe_wine = pd.read_csv('winequality-red.csv', sep=';')
```

```python
dframe_wine.head() # quality=評価
```

```python
dframe_wine['alcohol'].mean() # 平均
```

```python
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
wino = dframe_wine.groupby('quality')
```

```python
wino.describe()
```

```python
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```

```python
wino.agg('median')
```
