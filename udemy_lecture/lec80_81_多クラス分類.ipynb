多クラス分類

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris= load_iris() # irisデータ取得
```

```python
X = iris.data
```

```python
Y = iris.target
```

```python
iris_data = DataFrame(X, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
```

```python
iris_target = DataFrame(Y, columns=['Species'])
```

```python
# Yの値をあやめの名前に変換
def flower(num):
    if num == 0:
        return 'Setosa'
    elif num == 1:
        return 'Veriscolour'
    else:
        return 'Virginica'
```

```python
iris_target['Species'] = iris_target['Species'].apply(flower)
```

```python
iris = pd.concat([iris_data, iris_target], axis=1)
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue = 'Species', size =2)   #pairplot:各変数間の関係を描写
```

```python
plt.figure(figsize=(12, 4))
sns.countplot('Petal Length', data=iris, hue = 'Species')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.cross_validation import train_test_split
```

```python
logreg = LogisticRegression() # インスタンス生成
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state=3) # test_sizeはテストに回すデータの割合 random_stateはseed値
```

```python
logreg.fit(X_train, Y_train)
```

```python
from sklearn import metrics
```

```python
Y_pred = logreg.predict(X_test) # 予測実行
```

```python
Y_pred
```

```python
metrics.accuracy_score(Y_test, Y_pred) # 精度確認。93%くらい
```

```python
# k近傍法を実行
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=6) # n_neighbors:近傍を何個とるか？
```

```python
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test) # 予測
```

```python
metrics.accuracy_score(Y_test, Y_pred)
```

```python
# k近傍法は、Kの数で正解が変わってくる。
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
metrics.accuracy_score(Y_test, Y_pred)
```

```python
k_range = range(1, 90)
accuracy = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accuracy.append(metrics.accuracy_score(Y_test, Y_pred))
```

```python
plt.plot(k_range, accuracy)
```
