```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot('total_bill', 'tip', tips)  # lm:lenier regression X軸にtotal_bill, Y軸にtipを取る
```

```python
sns.lmplot('total_bill', 'tip', tips, scatter_kws={'marker': 'o', 'color' : 'indianred'}, line_kws={'linewidth': 1, 'color':'blue'})
```

```python
sns.lmplot('total_bill', 'tip', tips, order = 4, scatter_kws={'marker': 'o', 'color' : 'indianred'}, line_kws={'linewidth': 1, 'color':'blue'})
# orderで次元数を指定できる
```

```python
sns.lmplot('total_bill', 'tip', tips, fit_reg=False) # fit_reg=Falseで散布図だけ描写
```

```python
tips['tip_pect'] = 100 *(tips['tip']/ tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size', 'tip_pect', tips)
```

```python
sns.lmplot('size', 'tip_pect', tips, x_jitter = 0.2)  # x_jitter:ゆらぎ幅の指定
```

```python
sns.lmplot('size', 'tip_pect', tips, x_estimator=np.mean)
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue='sex', markers=['x','o']) # hue:色調・傾向 性別で傾向をみる
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue='day') 
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, lowess=True , line_kws={'color':'black'})
```

```python
sns.regplot('total_bill', 'tip_pect', tips)
```

```python
fig, (axis1, axis2) = plt.subplots(1, 2, sharey=True) #1行2列
sns.regplot('total_bill', 'tip_pect', tips, ax=axis1)
sns.violinplot(y="tip_pect", x='size', data=tips.sort('size'), ax=axis2)
```
