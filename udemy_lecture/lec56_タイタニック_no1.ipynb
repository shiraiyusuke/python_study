```python
# 16歳未満を子供とします。
def male_female_child(passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
    
# personという新しい列を追加。applyは、1行ずつに関数を適応する。
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
fig = sns.FacetGrid(titanic_df, hue='Sex', aspect=4) 
# FacetGridは、一つのプロットを
# aspectはアスペクト比を指定。横長とかにできる。
fig.map(sns.kdeplot, 'Age', shade=True) # カーネル密度推定
oldest = titanic_df['Age'].max()
fig.set(xlim=(0, oldest)) # 図を年齢の幅で抑える。
fig.add_legend()
```

```python
# 客室ランクごとに、カーネル密度推定
fig = sns.FacetGrid(titanic_df, hue='Pclass', aspect=4) 
fig.map(sns.kdeplot, 'Age', shade=True) # カーネル密度推定
oldest = titanic_df['Age'].max()
fig.set(xlim=(0, oldest)) # 図を年齢の幅で抑える。
fig.add_legend()

# ランクが低いと、年齢層が低くなる傾向が見て取れる。
```

```python
# NaNの行を取り除く
deck = titanic_df['Cabin'].dropna()
```

```python
deck
```

```python
# cabinの先頭1文字(A〜C)が階を表すので、それを抽出
levels = []
for level in deck:
    levels.append(level[0])
```

```python
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
```

```python
cabin_df
```

```python
sns.countplot('Cabin', data=cabin_df, palette='winter_d', order = sorted(set(levels)))
# palette:文字列を指定することで、色合いを指定できる。 _dをつけると、ちょっと暗めになる。
```

```python
# Tだけ、アルファベットとして飛んでいて、外れ値っぽいぞ！　取り除こう
cabin_df = cabin_df[cabin_df.Cabin != 'T']
```

```python
sns.countplot('Cabin', data=cabin_df, palette='summer', order = sorted(set(cabin_df.Cabin)))
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no', 1: 'yes'})
```

```python
sns.factorplot('Pclass', 'Survived', data=titanic_df, order=[1, 2, 3]) # factorplot:PclassとSurvivedの関係をplotする。
```

```python
sns.factorplot('Pclass', 'Survived', hue='person', data = titanic_df, order = [1, 2, 3], aspect =2)
# 男性の生存率が低い。けど、ランク3の客室は全体的に低い
```

```python
sns.lmplot('Age', 'Survived', data= titanic_df)  # lmplot:回帰直線を引く
```

```python
sns.lmplot('Age', 'Survived', hue = 'Pclass', data= titanic_df, palette='winter', hue_order=[1, 2, 3])
```

```python
generations = [10, 20, 40, 60, 80]
sns.lmplot('Age', 'Survived', hue = 'Pclass', data= titanic_df, palette='winter', hue_order=[1, 2, 3], x_bins=generations) # xbinsをつけると標準偏差も描写
```

```python
sns.lmplot('Age', 'Survived', hue = 'Sex', data= titanic_df, palette='winter', x_bins=generations)
# 女性は年齢が上がると、生存率が高くなっている傾向にあった。
```
