```python
import numpy as np
from numpy.random import randn
from scipy import stats
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100) + 2
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
sns.boxplot(data = [data1, data2])
```

```python
sns.boxplot(data = [data1, data2], whis=np.inf) # whis:ひげ np.inf:∞　外れ値を含んでひげが伸びる
```

```python
sns.boxplot(data=[data1, data2], orient='h')
```

```python
sns.violinplot(data = [data1, data2])
```

```python
data1 = stats.norm(0, 5).rvs(100)
```

```python
data2 = np.concatenate([stats.gamma(5).rvs(50) - 1, -1 * stats.gamma(5).rvs(50)])
```

```python
sns.boxplot(data = [data1, data2])
```

```python
sns.violinplot(data = [data1, data2]) # violinplotだと、data2には山が二つあることが見て取れる。
```

```python
sns.violinplot(data = data2, bw = 0.01)
```

```python
sns.violinplot(data = data1, inner='stick')
```
