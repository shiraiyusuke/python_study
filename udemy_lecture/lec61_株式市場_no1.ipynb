## 1年後の株価を予測したい。　
####ブラウン運動モデル（ランダムな動き） ΔS/S = μΔt + σ ε √Δt     μΔt: ドリフト      σ ε √Δt:　ショックと呼ぶ
+ S  株価 
+ μ  平均
+ Δt  時間間隔
+ σ  μの標準偏差 
+ ε:乱数

### εが乱数なので、ここにランダム性が生まれる。乱数を使ったシミュレーションのことをモンテカルロ法と呼ぶ。

```python
days = 365
dt = 1/ days # Δt
mu = rets.mean()['GOOG'] # μ
sigma = rets.std()['GOOG']  # σ
```

```python
# シミュレーション実行関数
def stock_monte_carlo(start_price, days, mu, sigma):
    price = np.zeros(days)
    price[0] = start_price
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    for x in range(1, days):
        shock[x] = np.random.normal(loc = mu*dt, scale = sigma * np.sqrt(dt)) # loc: 平均 scale:標準偏差
        drift[x] = mu * dt
        price[x] = price[x - 1] + (price[x - 1] * (drift[x] +shock[x]))    # (price[x - 1] * (drift[x] +shock[x]))　これが、S(μΔt+σϵ√Δt)を示す。
    return price
```

```python
GOOG.head()
```

```python
start_price = GOOG.iloc[0, 5] # GOOGの0行目の5列目のデータ取得

for run in range(5):
    plt.plot(stock_monte_carlo(start_price, days, mu, sigma))

plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis')
```

```python
loop_cnt = 100
test_array = np.zeros(loop_cnt)
for i in range(loop_cnt):
    test_array[i] = np.random.normal(loc = 0, scale = 1)

sns.distplot(test_array, bins =20)
```

```python
GOOG.iloc[0, 4]
```

```python
runs= 10000
simulations = np.zeros(runs)
np.set_printoptions(threshold = 5) # 表示のオプション
for run in range(runs):
    simulations[run] = stock_monte_carlo(start_price, days, mu, sigma)[days -1]
```

```python
plt.hist(simulations, bins = 200)
```

```python
q = np.percentile(simulations, 1)
plt.hist(simulations, bins = 200)

plt.figtext(0.7, 0.8, s='Start price: {:0.2f}'.format(start_price))
plt.figtext(0.7, 0.7, 'mean final price: {:0.2f}'.format(simulations.mean()))
plt.figtext(0.7, 0.6, 'VaR(0.99): {:0.2f}'.format(start_price -q))
plt.figtext(0.15, 0.6, 'q(0.99): {:0.2f}'.format(q))

plt.axvline(x = q, linewidth = 4, color = 'r') # axvline:
```

### この結果から、「537ドルで1株買って、99%の信頼区間で22.09ドルより損することはないですよ」ってことが言える。
