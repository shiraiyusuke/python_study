```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
```

```python
plt.hist(dataset, alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)

x_min = dataset.min() - 2
x_max = dataset.max() + 2
x_axis = np.linspace(x_min, x_max, 100)

bandwidth = ((4*dataset.std()**5/(3*len(dataset)))) **0.2

kernel_list =[]
for data_point in dataset:
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)    # scipyのstats
    kernel_list.append(kernel)
    
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    plt.plot(x_axis, kernel, color = 'gray', alpha = 0.5)
plt.ylim(0, 1)
```

```python
sum_of_kde = np.sum(kernel_list, axis = 0)

fig = plt.plot(x_axis, sum_of_kde, color = 'indianred')
sns.rugplot(dataset)
plt.yticks([]) # y軸のdicsを消す
plt.suptitle('Sum of the Basis Functions')
```

```python
sns.kdeplot(dataset) # 上と同じグラフがsnsで簡単にかけちゃう
```

```python
sns.rugplot(dataset, color = 'black')
for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(dataset, bw=bw, label=bw)
```

```python
kernel_options = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']
for kern in kernel_options:
    sns.kdeplot(dataset, kernel = kern, label = kern)
```

```python
sns.kdeplot(dataset, vertical=True) # 軸をずらすことも可能
```

```python
plt.hist(dataset, cumulative=True) # 累積関数
```

```python
sns.kdeplot(dataset, cumulative=True)
```
