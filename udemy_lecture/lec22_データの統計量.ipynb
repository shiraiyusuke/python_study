```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
arr = np.array([[1, 2, np.nan], [np.nan, 3, 4,]])
```

```python
arr
```

```python
dframe1 = DataFrame(arr, index = ['A', 'B'], columns = ['One', 'Two', 'Three'])
```

```python
dframe1
```

```python
dframe1.sum() # 列ごとにsumを計算
```

```python
dframe1.sum(axis =1) # 行方向はaxis=1を指定
```

```python
dframe1.min() # 縦方向に最小値
```

```python
dframe1.min(axis = 1)
```

```python
dframe1.idxmin()  #どのindexの値が最小かをみる
```

```python
dframe1.cumsum() # 累積の計算
```

```python
dframe1.describe()  # 統計情報を出してくれる
```

```python
import pandas.io.data as pdweb
import datetime
```

```python
# アメリカの株価のデータ取得。
prices = pdweb.get_data_yahoo(['CVX', 'XOM', 'BP'],
                              start=datetime.datetime(2010, 1, 1),
                              end=datetime.datetime(2013, 1,1))['Adj Close']
```

```python
prices
```

```python
prices.head()
```

```python
rets = prices.pct_change() # 変化率
```

```python
rets.head()
```

```python
%matplotlib inline
# jyupterにmatplotlibを使う宣言
```

```python
prices.plot()
```

```python
rets.corr() # 相関係数
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(rets.corr())
```
